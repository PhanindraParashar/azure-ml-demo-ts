```python
import os 
import pandas as pd

def pipe_reset_index(df: pd.DataFrame) -> pd.DataFrame:
    df.index = range(len(df))
    return df
```

```python
df = pd.read_csv('../data/full_data.csv')
```

```python
df_holdout = df[df['date'] >= '2017-01-01'].pipe(pipe_reset_index)
df_holdout.to_csv('../data/holdout.csv', index=False)

df_train = df[df['date'] < '2017-01-01'].pipe(pipe_reset_index)
df_train.to_csv('../data/train_test.csv', index=False)
```
